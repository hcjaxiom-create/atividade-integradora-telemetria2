## 1.4 Análise energética

A análise energética foi realizada com o objetivo de estimar a autonomia inicial da nave Asterion V antes da decolagem. Para isso, foram considerados a capacidade total do sistema de energia, o percentual de carga disponível no momento da verificação, o consumo estimado durante a decolagem e as perdas energéticas do sistema.

Adotou-se como capacidade total o valor de **1000 kWh**, com **84% de carga atual**, resultando em uma disponibilidade inicial de energia proporcional ao nível armazenado. Em seguida, foram consideradas **perdas energéticas de 5%**, representando dissipações naturais do sistema. Após esse desconto, foi subtraído o **consumo estimado de 250 kWh** referente ao procedimento de decolagem.

Com base nesses cálculos, foi possível determinar a energia útil da nave e sua autonomia inicial após a etapa de lançamento.

```python
# 1.4 Análise energética

capacidade_total = 1000      # kWh
carga_atual = 84             # %
consumo_decolagem = 250      # kWh
perdas_percentuais = 5       # %

energia_disponivel = capacidade_total * (carga_atual / 100)
perdas_energeticas = energia_disponivel * (perdas_percentuais / 100)
energia_util = energia_disponivel - perdas_energeticas
autonomia_inicial = energia_util - consumo_decolagem

print("=== ANÁLISE ENERGÉTICA ===")
print(f"Capacidade total: {capacidade_total} kWh")
print(f"Carga atual: {carga_atual}%")
print(f"Consumo estimado na decolagem: {consumo_decolagem} kWh")
print(f"Perdas energéticas: {perdas_percentuais}%\n")

print(f"Energia disponível inicial: {energia_disponivel:.2f} kWh")
print(f"Perdas energéticas: {perdas_energeticas:.2f} kWh")
print(f"Energia útil após perdas: {energia_util:.2f} kWh")
print(f"Autonomia inicial após decolagem: {autonomia_inicial:.2f} kWh")
```
